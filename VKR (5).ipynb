```python
# Загрузка нескольких CSV файлов
df1 = pd.read_csv('/content/drive/MyDrive/vkr/Air Conditioners.csv')
df2 = pd.read_csv('/content/drive/MyDrive/vkr/All Appliances.csv')
df3 = pd.read_csv('/content/drive/MyDrive/vkr/All Books.csv')
df4 = pd.read_csv('/content/drive/MyDrive/vkr/All Car and Motorbike Products.csv')
df5 = pd.read_csv('/content/drive/MyDrive/vkr/All Electronics.csv')
df6 = pd.read_csv('/content/drive/MyDrive/vkr/All English.csv')
df7 = pd.read_csv('/content/drive/MyDrive/vkr/All Exercise and Fitness.csv')
df8 = pd.read_csv('/content/drive/MyDrive/vkr/All Grocery and Gourmet Foods.csv')
df9 = pd.read_csv('/content/drive/MyDrive/vkr/All Hindi.csv')
df10 = pd.read_csv('/content/drive/MyDrive/vkr/All Home and Kitchen.csv')
df11 = pd.read_csv('/content/drive/MyDrive/vkr/All Movies and TV Shows.csv')
df12 = pd.read_csv('/content/drive/MyDrive/vkr/All Music.csv')
df13 = pd.read_csv('/content/drive/MyDrive/vkr/All Pet Supplies.csv')
df14 = pd.read_csv('/content/drive/MyDrive/vkr/All Sports Fitness and Outdoors.csv')
df15 = pd.read_csv('/content/drive/MyDrive/vkr/All Video Games.csv')
df16 = pd.read_csv('/content/drive/MyDrive/vkr/Amazon Fashion.csv')
df17 = pd.read_csv('/content/drive/MyDrive/vkr/Amazon Pharmacy.csv')
df18 = pd.read_csv('/content/drive/MyDrive/vkr/Amazon-Products.csv')
df19 = pd.read_csv('/content/drive/MyDrive/vkr/Baby Bath Skin and Grooming.csv')
df20 = pd.read_csv('/content/drive/MyDrive/vkr/Baby Fashion.csv')
df21 = pd.read_csv('/content/drive/MyDrive/vkr/Baby Products.csv')
df22 = pd.read_csv('/content/drive/MyDrive/vkr/Backpacks.csv')


# Объединение датасетов (пример для вертикального объединения)
combined_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14,
                         df15, df16, df17, df18, df19, df20, df21, df22], axis=0)
```

```python
combined_df
```

```python
# Получим уникальные значения
main_categories = combined_df['main_category'].unique()
sub_categories = combined_df['sub_category'].unique()

# Построим словари вручную или через переводчик
main_category_dict = {cat: translator.translate(cat, dest='ru').text for cat in tqdm(main_categories)}
sub_category_dict = {cat: translator.translate(cat, dest='ru').text for cat in tqdm(sub_categories)}

# Применим перевод
combined_df['main_category_ru'] = combined_df['main_category'].map(main_category_dict)
combined_df['sub_category_ru'] = combined_df['sub_category'].map(sub_category_dict)
```

```python
combined_df.head(5)
```

```python
# Список ненужных столбцов (замени на актуальные названия в твоём случае)
columns_to_drop = [
    'image', 'main_category',
    'sub_category', 'Unnamed: 0', 'no_of_ratings', 'discount_price'
]

# Удалим столбцы, если они есть
df = combined_df.drop(columns=[col for col in columns_to_drop if col in combined_df.columns])

# Очистка actual_price: удалим символы валют и запятые, приведём к float
df['actual_price'] = (
    df['actual_price']
    .astype(str)
    .replace(r'[₹$,]', '', regex=True)
    .replace(r',', '', regex=True)
    .astype(str)
    .str.strip()
)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Очистка ratings: убираем всё нечисловое
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Удалим строки с пропущенными значениями в ключевых столбцах
df = df.dropna(subset=['actual_price', 'ratings'])
```
